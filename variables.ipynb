# Variable in Python



**A variable is a name that holds a value. Variables are fundamental to programming because they allow you to store, modify, and reference data within your code.**

**Defining Variables**

You create a variable by assigning it a value using the = sign

**variable_name = value**


x = 10

name = "Alice"


**Naming Rules for Variables in Python:**



*  Start with a Letter or Underscore (_):
*  Variable names must begin with a letter (a–z, A–Z) or an underscore (_).





_name = "Alice"  # valid

age = 25         # valid


*  Cannot Start with a Number: Variables cannot start with a number

1name = "Alice"  # invalid


*  Only Letters, Numbers, and Underscores: Variable names can only contain letters, numbers, and underscores.

age_1 = 30       # valid

age! = 30        # invalid


*  Case-Sensitive: Python variables are case-sensitive, meaning Age and age are considered different variables.




Age = 30         # valid

age = 25         # different from `Age`


*  Avoid Reserved Words: Certain words are reserved by Python (like class, for, return) and cannot be used as variable names.

**Best Practices for Naming Variables**

*  Descriptive Names: Use descriptive names that indicate what the variable holds. For instance, total_sales is more descriptive than x.

*  Use Snake Case: By convention, use snake case for variable names, which means words are separated by underscores (_). For example:
total_sales = 1000
user_age = 25


*  Avoid Special Characters and Spaces: Only underscores are allowed, not spaces or special symbols (like @, !, etc.).


**Reassigning Variables**

*  Variables in Python can be reassigned to different values, even of a different type:

x = 10       # initially an integer

x = "Alice"  # now a string


**Variables in Expressions**

*  Variables can be used in expressions, combined with other variables, or passed into functions.

a = 5

b = 10

sum = a + b   # sum is 15

# Mini Project: Simple Budget Tracker

```python
# Step 1: Initial budget
total_budget = 1000

# Step 2: Input expenses
food_expense = int(input("Enter your food expense: "))
entertainment_expense = int(input("Enter your entertainment expense: "))
transport_expense = int(input("Enter your transport expense: "))

# Step 3: Calculate total expenses and remaining budget
total_expenses = food_expense + entertainment_expense + transport_expense
remaining_budget = total_budget - total_expenses

# Step 4: Display results
print("\nExpense Summary:")
print(f"Food: ${food_expense}")
print(f"Entertainment: ${entertainment_expense}")
print(f"Transport: ${transport_expense}")
print(f"Total Expenses: ${total_expenses}")
print(f"Remaining Budget: ${remaining_budget}")
```
